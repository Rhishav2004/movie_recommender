```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head(1)
```

```python
movies['cast'][0]
```

```python
def convert2(obj):
    L=[]
    c=0
    for i in ast.literal_eval(obj):
        if c!=3:
            L.append(i['name'])
            c+=1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert2)
```

```python
movies['crew'][0]
```

In cast column , we only need to extract the name from the dictionary where job is 'Director'

```python
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

The overview column contains Lists , so we need to change it to List so that it can be appended later

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

```python
movies.head(2)
```

We need to remove the spaces
'Sam Worthington' -> 'SamWorthington'

```python
def remove_space(L):      
    return [i.replace(' ', '') for i in L]
```

```python
movies['genres'] = movies['genres'].apply(remove_space)
movies['keywords'] = movies['keywords'].apply(remove_space)
movies['overview'] = movies['overview'].apply(remove_space)
movies['cast'] = movies['cast'].apply(remove_space)
movies['crew'] = movies['crew'].apply(remove_space)
```

```python
movies['tags']= movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head(1)
```

```python
new_df = movies[['movie_id', 'title' , 'tags']]
```

```python
new_df.head(2)
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
```

```python
new_df['tags'][0]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
```

```python
new_df.head()
```

['loving' ,'love', 'loved'] these words are counted differently, thus causing problem
we need to solve this such that 

['actions' , 'action '] -> ['action', 'action']
['loving' ,'love', 'loved'] ->['love','love', 'love'] 

for this we nltk

```python
import nltk
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
ps.stem('loving')
```

```python
ps.stem('loved')
```

```python
def stem(text):
    y=[]

    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000 , stop_words = 'english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
list(enumerate(similarity[0]))
```

```python
sorted(list(enumerate(similarity[0])), reverse = True)
```

 here we see sorting is done on the basis of indexes while it has to be done on the basis of similiarities

```python
sorted(list(enumerate(similarity[0])), reverse = True , key=lambda x: x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)), reverse = True , key=lambda x: x[1])[1:6]
    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```

```python
recommend('Batman')
```
